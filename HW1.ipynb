## Python Skills

Demonstrate your ability to:
- Import modules
- Declare lists and NumPy arrays
- Declare basic Python functions
- Make simple Matplotlib figures and save them as PNGs

```python
# Import modules
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Declare a list and a NumPy array
my_list = [1, 2, 3, 4, 5]
my_array = np.array([1, 2, 3, 4, 5])
print("List:", my_list)
print("NumPy Array:", my_array)
```

```python
# Declare a basic Python function
def square(x):
    return x ** 2

print("Square of 3:", square(3))
```

```python
# Make a simple Matplotlib figure and save as PNG
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.title("Sine Wave")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.savefig("sine_wave.png")
plt.show()
```
